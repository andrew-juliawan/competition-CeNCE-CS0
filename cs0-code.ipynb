## Q2
#### Estimation of pi using Leibniz Formula.

```python
def estimate_pi(k):
    sum = 0
    for i in range (1, k+1):
        sum += ((-1)**i)/(2*i+1)
    return 4*sum
```

## Q3
#### Simulation of water volume in a tank over time.

```python
def sim(cap, d):
    # your code here
    vol = 0
    for i in range(0, d+1):
        vol += (d%67)*100
        if vol > cap:
            vol = cap
        vol -= (2*(d%25))**2
        if vol < 0:
            vol = 0
        if i%3 == 0:
            vol += 50
        if vol < 0:
            vol = 0
        
    vol = vol - vol*0.01
    
    return round(vol, 3)
```
